## Question 1
**What is hypothesis testing in statistics?**

**Answer:**

Hypothesis testing is a formal procedure for making decisions about population parameters using sample data. Formulate H0 and H1, choose test statistic, compute p-value or compare to critical value, then make a decision.

## Question 2
**What is the null hypothesis, and how does it differ from the alternative hypothesis?**

**Answer:**

H0 is the baseline claim (no effect), H1 is the claim we test for (effect exists). Rejecting H0 supports H1.

## Question 3
**Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.**

**Answer:**

Significance level alpha is the threshold for Type I error; p-value <= alpha leads to rejecting H0.

## Question 4
**What are Type I and Type II errors? Give examples of each.**

**Answer:**

Type I: false positive — rejecting true H0. Type II: false negative — not rejecting false H0. Examples: drug trial errors.

## Question 5
**What is the difference between a Z-test and a T-test? Explain when to use each.**

**Answer:**

Z-test uses known sigma and normal distribution; t-test uses sample s and t-distribution, especially for small samples.

## Question 6
**Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram?**

```python

import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
samples = np.random.binomial(n=10, p=0.5, size=1000)
counts = [int((samples==k).sum()) for k in range(11)]
print("Counts 0..10:", counts)
plt.bar(range(11), counts)
plt.xlabel("Successes out of 10"); plt.ylabel("Frequency"); plt.title("Binomial(n=10,p=0.5) - 1000 draws")
plt.show()
```

## Question 7
**Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.**

**sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]**

```python

import statistics, math
from math import sqrt
from scipy import stats

sample_data = [49.1,50.2,51.0,48.7,50.5,49.8,50.3,50.7,50.2,49.6,
               50.1,49.9,50.8,50.4,48.9,50.6,50.0,49.7,50.2,49.5,
               50.1,50.3,50.4,50.5,50.0,50.7,49.3,49.8,50.2,50.9,
               50.3,50.4,50.0,49.7,50.5,49.9]

mu0 = 50
n = len(sample_data)
xbar = statistics.mean(sample_data)
s = statistics.stdev(sample_data)
z_stat = (xbar - mu0) / (s / math.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("n =", n)
print("Sample mean =", xbar)
print("Sample std (s) =", s)
print("Approx Z-statistic =", z_stat)
print("Two-sided p-value (approx) =", p_value)
```

## Question 8
**Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.**

```python

import numpy as np, statistics, math
import matplotlib.pyplot as plt
from scipy import stats as st

np.random.seed(0)
data = np.random.normal(loc=100, scale=15, size=200)
n = len(data)
xbar = np.mean(data)
s = np.std(data, ddof=1)
se = s / math.sqrt(n)
t_crit = st.t.ppf(0.975, df=n-1)
ci = (xbar - t_crit*se, xbar + t_crit*se)

print("Sample mean:", xbar)
print("Sample std (s):", s)
print("95% CI:", ci)

plt.plot(data, marker='.', linestyle='none')
plt.axhline(xbar, linestyle='--')
plt.title("Simulated normal data (n=200)")
plt.show()
```

## Question 9
**Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.**

```python

import statistics, math
import matplotlib.pyplot as plt

def z_scores(data):
    mu = statistics.mean(data)
    sigma = statistics.pstdev(data)
    return [(x - mu) / sigma for x in data]

data_demo = [48,50,55,60,52,49,51,47,53,58,50,52,54,49,46]
zs = z_scores(data_demo)
print("Z-scores:", [round(z,3) for z in zs])

plt.hist(zs, bins=8, edgecolor='black')
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score"); plt.ylabel("Frequency")
plt.show()
```
