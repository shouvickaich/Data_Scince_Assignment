## Question 1
**What is Deep Learning? Briefly describe how it evolved and how it differs from traditional machine learning.**

**Answer:**
Deep Learning is a subset of machine learning that uses artificial neural networks with multiple hidden layers to automatically learn hierarchical representations of data. It evolved from early neural networks like perceptrons and gained momentum due to increased computational power (GPUs), availability of big data, and improved algorithms such as backpropagation.

Traditional machine learning relies heavily on manual feature engineering, whereas deep learning automatically extracts features from raw data. Deep learning models are especially powerful for unstructured data such as images, audio, and text.

## Question 2
**Explain the basic architecture and functioning of a Perceptron. What are its limitations?**

**Answer:**
A perceptron consists of input nodes, weights, a bias, a summation function, and an activation function. Inputs are multiplied by weights, summed, bias is added, and the result is passed through an activation function to produce output.

**Limitations:**
- Can only solve linearly separable problems
- Cannot handle XOR problem
- No hidden layers

## Question 3
**Describe the purpose of activation function. Compare Sigmoid, ReLU, and Tanh.**

**Answer:**
Activation functions introduce non-linearity.

- **Sigmoid:** Output (0,1), suffers from vanishing gradient
- **Tanh:** Output (-1,1), zero-centered
- **ReLU:** f(x)=max(0,x), faster convergence but dead neuron problem

## Question 4
**Difference between Loss and Cost function with examples.**

**Answer:**
Loss function measures error for a single sample (e.g., Mean Squared Error). Cost function is the average loss over the dataset.

**Answer:**
Optimizers update weights to minimize loss.

- Gradient Descent: simple but slow
- Adam: adaptive learning rates, fast convergence
- RMSprop: handles non-stationary objectives

##Question 6:

 **Write a Python program to implement a single-layer perceptron from scratch using NumPy to solve the logical AND gate.**

```python
import numpy as np

# AND Gate Perceptron
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])

w = np.zeros(2)
b = 0
lr = 0.1

def step(x): return 1 if x >= 0 else 0

for _ in range(10):
    for i in range(len(X)):
        y_pred = step(np.dot(X[i], w) + b)
        w += lr * (y[i] - y_pred) * X[i]
        b += lr * (y[i] - y_pred)

print("Weights:", w, "Bias:", b)
```

##Question 7:

**Implement and visualize Sigmoid, ReLU, and Tanh activation functions using Matplotlib.**

```python

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)

sigmoid = 1/(1+np.exp(-x))
relu = np.maximum(0,x)
tanh = np.tanh(x)

plt.plot(x,sigmoid,label='Sigmoid')
plt.plot(x,relu,label='ReLU')
plt.plot(x,tanh,label='Tanh')
plt.legend()
plt.show()
```

##Question 8:

**Use Keras to build and train a simple multilayer neural network on the MNIST digits dataset. Print the training accuracy.**

```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

(X_train,y_train),(X_test,y_test)=mnist.load_data()
X_train=X_train/255.0
y_train=to_categorical(y_train)

model=Sequential([Flatten(input_shape=(28,28)),Dense(128,activation='relu'),Dense(10,activation='softmax')])
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
history=model.fit(X_train,y_train,epochs=5,batch_size=32)
print('Training Accuracy:',history.history['accuracy'][-1])
```

## Question 9
**Visualize the loss and accuracy curves for a neural network model trained on the Fashion MNIST dataset. Interpret the training behavior.**


```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import fashion_mnist

# Load dataset
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# Normalize data
X_train = X_train / 255.0
X_test = X_test / 255.0

# Build model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train model
history = model.fit(X_train, y_train, epochs=10, validation_split=0.2)

# Plot loss and accuracy
plt.figure()
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.show()

plt.figure()
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.legend()
plt.show()
```

# **Interpretation of Training Behavior:**

*   Training loss decreases steadily, indicating the model is learning patterns from the data.

*   Training accuracy increases, showing improved classification performance.

*   Validation loss closely follows training loss, which means the model generalizes well.

*   If validation loss increases while training loss decreases, it indicates overfitting.

*   In this case, balanced curves suggest a well-trained and stable model.

**You are working on a project for a bank that wants to automatically detect fraudulent transactions. The dataset is large, imbalanced, and contains structured features like transaction amount, merchant ID, and customer location. The goal is to classify each transaction as fraudulent or legitimate.
Explain your real-time data science workflow:

● Which activation function and loss function would you use, and why?

● How would you train and evaluate the model, considering class imbalance?

● Which optimizer would be suitable, and how would you prevent overfitting?**



# **1. Deep Learning Model Design**

For fraud detection, a Multilayer Neural Network (MLP) is more suitable than a single perceptron because:

*   Fraud patterns are non-linear and complex

*   Relationships between features (amount, merchant, location, time) are not linearly separable

**Model structure:**

*   **Input layer**: Structured numerical and encoded categorical features

*   **Hidden layers**: 2–3 dense layers to learn complex patterns

*   **Output layer**: Single neuron for binary classification (fraud / legitimate)

A single-layer perceptron would be insufficient because it can only handle linearly separable data.


---


# **2. Activation Function and Loss Function**

Hidden layers: ReLU (Rectified Linear Unit)

*   Faster training

*   Avoids vanishing gradient problem

*   Works well for deep networks

Output layer: Sigmoid

*   Produces probability between 0 and 1

*   Ideal for binary classification

Loss function: Binary Crossentropy

*   Measures the difference between predicted probability and actual class

*   Standard and effective for fraud detection problems


---



# **3. Training and Evaluation (Handling Class Imbalance)**

Fraud datasets are highly imbalanced (very few fraud cases).

Techniques to handle imbalance:

*   Apply class weights to penalize misclassification of fraud cases

*   Use oversampling (e.g., SMOTE) or undersampling

*   Use stratified train-test split

Evaluation metrics (accuracy alone is misleading):

*   Precision

*   Recall (very important for fraud detection)

*   F1-Score

*   ROC-AUC

The goal is to maximize recall so that fraudulent transactions are not missed.


---



# **4. Optimizer and Overfitting Prevention**

Optimizer: Adam

*   Adaptive learning rate

*   Fast convergence on large datasets

*   Works well with sparse and noisy data

Overfitting prevention techniques:

*   Dropout layers to randomly deactivate neurons

*   Early stopping based on validation loss

*   L2 regularization to penalize large weights

*   Cross-validation

#Conclusion

A multilayer neural network with ReLU activations, sigmoid output, binary crossentropy loss, and Adam optimizer, combined with imbalance-handling techniques and regularization, provides a robust, scalable, and real-time fraud detection system suitable for banking applications.
