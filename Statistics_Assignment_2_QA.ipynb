**What is a binomial distribution, and how is it used in probability?**

**Answer:**

Models number of successes in n independent Bernoulli trials with probability p. PMF: C(n,k)p^k(1-p)^(n-k).

## Question 6

**Answer:**

CLT: sampling distribution of sample mean approaches normal for large n, enabling inference.

## Question 7

**Answer:**

## Question 9
**Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution?**

**(Include your Python code and output in the code box below.)**


```python

import numpy as np
import statistics, math
import matplotlib.pyplot as plt

np.random.seed(0)
samples = np.random.normal(50, 5, 1000)
print("Sample mean:", np.mean(samples))
print("Sample std (ddof=1):", np.std(samples, ddof=1))

plt.hist(samples, bins=30, edgecolor='black')
plt.title("Histogram of 1000 samples from N(50,5^2)")
plt.xlabel("Value"); plt.ylabel("Frequency")
plt.show()
```


## Question 10
**You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend.**

**daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
235, 260, 245, 250, 225, 270, 265, 255, 250, 260]**


  **Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval.**


**Write the Python code to compute the mean sales and its confidence interval.**

**(Include your Python code and output in the code box below.)**


```python

import statistics, math
daily_sales = [220,245,210,265,230,250,260,275,240,255,235,260,245,250,225,270,265,255,250,260]
n = len(daily_sales)
mean_sales = statistics.mean(daily_sales)
s = statistics.stdev(daily_sales)

# Attempt to use scipy.t; fallback to t_crit for df=19 (~2.093)
try:
    from scipy import stats
    t_crit = stats.t.ppf(0.975, df=n-1)
except Exception:
    t_crit = 2.093

se = s / math.sqrt(n)
ci = (mean_sales - t_crit*se, mean_sales + t_crit*se)

print("n =", n)
print("Sample mean =", mean_sales)
print("Sample std (s) =", s)
print("t_crit (95%) =", t_crit)
print("95% CI for mean =", (round(ci[0],2), round(ci[1],2)))
```
